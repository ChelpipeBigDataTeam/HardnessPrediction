```python
df = df.reset_index()
```

```python
columns = ['index', 'Литературный источник']
df = df.drop(columns, 1)
```

```python
df
```

```python
indexes = df[df['Сталь'].notnull()].index
indexes.size
```

```python
tables = []
```

```python
for i in range (indexes.size):
    if i == indexes.size-1:
        tmp = df.iloc[indexes[i]:indexes[i]+10,:]
        tmp = np.split(tmp, [21], axis=1)
        tables.append(tmp[0])
    else:
        tmp = df.iloc[indexes[i]:indexes[i+1],:]
        tmp = np.split(tmp, [21], axis=1)
        tables.append(tmp[0])
    
    tmp1 = tmp[1].dropna(how='all').dropna(axis=1, how='all')
    tmp1['Параметры'] = tmp1['микро'].combine_first(tmp1['твердость'])
    cols = list(tmp1)
    cols.insert(0, cols.pop(cols.index('Параметры')))
    tmp1 = tmp1.ix[:, cols]
    columns = ['микро', 'твердость']
    tmp1 = tmp1.drop(columns, 1)
    if i > 1:
        tmp1 = tmp1.iloc[1:]
    tmp1 = tmp1.set_index('Параметры').T
    names = tmp1.columns.tolist()
    names[0]='V800-500 °C/c'
    tmp1.columns = names
    tables.append(tmp1)
```

```python
list = []
for i in range (0,len(tables),2):
    tables[i]['Сталь'][tables[i].index[1]] = tables[i]['Сталь'][tables[i].index[0]]
    tmp = tables[i].iloc[1:2].reset_index()
    if (tables[i+1].empty == False):
        size = tables[i+1].shape[0]
        columns = ['level_0','index']
        first = pd.concat([tmp]*size).reset_index().drop(columns,1)
        second = tables[i+1].reset_index().drop('index',1).dropna(how='all')
        result = pd.merge(first, second, left_index=True, right_index=True)
        list.append(result)
```

```python
prepared_data = pd.concat(list)
```
